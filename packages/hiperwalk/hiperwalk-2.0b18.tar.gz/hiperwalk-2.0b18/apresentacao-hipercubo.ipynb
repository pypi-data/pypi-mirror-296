# Busca no Hipercubo com 1 Vértice Marcado

## Importando bibliotecas necessárias

```python
import hiperwalk as hpw
import numpy as np
```

## Criando um hipercubo

```python
n = 12
g = hpw.Hypercube(n)
N = g.number_of_vertices()
```

## Configurando o passeio

```python
marked = [0]
num_marked = len(marked)
qw = hpw.Coined(g)
qw.set_marked({'-I': marked})
```

## Simulação

Da hipótese,

$$t_{opt} = \frac{\pi}{2\sqrt2} \sqrt{\frac{N}{|M|}}$$ .

```python
t_opt = int(np.pi/np.sqrt(8) * np.sqrt(N/len(qw.get_marked())))

states = qw.simulate(range=2*int(t_opt), state=qw.uniform_state())
p_succ = qw.success_probability(states)
```

## Gráfico da probabilidade de sucesso

```python
hpw.plot_success_probability(2*int(t_opt), p_succ)
```
