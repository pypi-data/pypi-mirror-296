```python
import ibis

ibis.options.interactive = True
ibis.options.repr.interactive.max_rows = 40

host = "clickpy-clickhouse.clickhouse.com"
port = 443
user = "play"
database = "pypi"

ch_con = ibis.clickhouse.connect(
    host=host,
    port=port,
    user=user,
    database=database,
)

t = ch_con.table("pypi_downloads_per_day_by_version_by_python").filter(
    ibis._["project"] == "ibis-framework"
)

agg = (
    t.mutate(version=ibis._["version"].split(".")[0])
    .group_by("date", "version", "python_minor")
    .agg(downloads=ibis._["count"].sum())
    .order_by(ibis.desc("date"), ibis.desc("downloads"))
)
agg
```
