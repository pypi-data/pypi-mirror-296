```python
import sempy_labs as labs
from sempy_labs import ConnectWarehouse
warehouse_name = ''
workspace_name = ''
```

### Run a SQL query against a Fabric warehouse

```python
with ConnectWarehouse(warehouse=warehouse_name, workspace=workspace_name) as sql:
    df = sql.query("SELECT * FROM Product")
```
